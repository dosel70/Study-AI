```python
import numpy as np
# 1~10까지의 요소를 ndarray에 담는다. -> arange 사용
ndarray = np.arange(start=1, stop=11)
print(ndarray)
print("="*30)
#------------------------------------------------------
# 3행 2열 list를 ndarray에 담고 각 요소는 순서대로 6,5,4,3,2,1 이다. -> arange ,slicing, sort 사용
ndarray = np.arange(start=1, stop=7)
ndarray3 = ndarray.reshape((3,2))
ndarray4 = ndarray3[::-1,::-1]
print(ndarray4)
print("="*30)
#------------------------------------------------------
# 4행 1열 ndarray, 순서대로 10,8,6,4를 담는다.
# Transpose : 전치, 행과 열을 바꿔준다. (ex)([list].T)
ndarray = np.arange(start=4, stop=11, step=2)
print(ndarray)
ndarray2 = ndarray.reshape((4,1))
ndarray3 = ndarray2[::-1,::-1]
print(ndarray3)
print("="*30)
#------------------------------------------------------
# 1 ~ 10까지 요소를 ndarray에 담고 각 요소에 5씩 더해준다.
ndarray = np.arange(start=1, stop=11)+5
print(ndarray)
print("="*30)
#------------------------------------------------------
# axis 0이 10인 shape이면서 모든 원소가 0, dtype은 int32인 ndarray 만들기
ndarray5 = np.zeros((10,), dtype=np.int32)
print(ndarray5.dtype, ndarray5)
print("="*30)
#------------------------------------------------------
# axis 0이 3(행), axis1이 4(열)인 shape 이면서 모든 원소가 1인 ndarray 만들기 -> ones 사용
ndarray6 = np.ones((3,4)) # 3행 4열
print(ndarray6)
print("="*30)
#------------------------------------------------------
# axis 0이 5, 각 요소가 0 ~ 4인 ndarray 만들기 
ndarray = np.arange(start=0, stop=5)
print(ndarray)
print("="*30)
#------------------------------------------------------
# 아래의 ndarray1이 주어졌을때 순서대로 문제를 해결한다.
# ndarray1 = np.arrange(start=0, stop=16)
ndarray1 = np.arange(start=0, stop=16)
# 2 Dimension, axis 1은 2로 변경
ndarray2 = ndarray1.reshape((-1,2))
print(ndarray2)
print("="*30)
#------------------------------------------------------
# 2 Dimension, axis 0은 8로 변경
ndarray3 = np.reshape(ndarray2, (8,-1))
print(ndarray3)
print("="*30)
#------------------------------------------------------
# 3 Dimension으로 변경
ndarray3 = ndarray3.reshape((2,2,4))
print("여기다!!!!!",ndarray3, ndarray3.ndim,"차원")
print("="*30)
#------------------------------------------------------
# ndarray3을 axis 1이 1인 2차원 ndarray로 변환 
ndarray2 = ndarray3.reshape((-1,1))
print(ndarray2) # <- 2차원
print(ndarray2.flatten())
# flatten -> 1차원으로 바꿔줌 
# ndarray3을 1 Dimension으로 변환 
print("="*30)
#------------------------------------------------------
# 1차원 ndarray 1~100까지 값중 97값 추출하기 
ndarray = np.arange(start=1, stop=101)
print(ndarray[-4])
print("="*30)

#------------------------------------------------------
# 아래의 ndarray1이 주어졌을 때 순서대로 문제를 해결한다.
ndarray1 = np.arange(start=1, stop=21)
print(ndarray1)

# 2차원 배열 axis0은 4, axis1은 5로 1~20 요소 값 중 15 값 추출하기
ndarray2 = ndarray1.reshape(4, 5)
print(ndarray2)

# 위치인덱싱을 사용하여 15 값 추출
ndarray2_result = ndarray2[2, 4]
print("15=",ndarray2_result)
print("="*30)
#---------------------------------------------------------

# 2 Dimension, axis0은 5로 변경 
ndarray3 = ndarray1.reshape(5,-1)
print(ndarray3)
print("="*30)
#------------------------------------------------------
# 2행의 4번째 숫자 출력 
print(ndarray3[1][3])
print("="*30)
#------------------------------------------------------
# 1 ~ 100 중 짝수만 출력 
ndarray = np.arange(start=2, stop=101, step=2)
print(ndarray)
#------------------------------------------------------
# 위에서 구한 짝수들을 axis 0 이 10인 2차원 배열로 변환 후 82 ~ 100까지 추출 
ndarray_change = ndarray.reshape(10,-1)
print(ndarray_change)
ndarray_result = ndarray_change[8:]
print(f"추출 값 =>{ndarray_result}")
print("="*30)
#-----------------------------------------------------
# 82 ~ 100까지 추출
ndarray_result = ndarray_change[8:]
print("추출 값:", ndarray_result)
print("="*30)

#(86, 100) 좌표 구하기
x_86 = ndarray_result[0][2] # 86
y_100 = ndarray_result[1][4] # 100
print(f"86의 좌표(x,y) : (0,2)") 
print(f"100의 좌표(x,y) : (1,4)")
print(x_86,y_100)
#-----------------------------------------------------
# # 아래의 ndarray1이 주어졌을 때 순서대로 문제를 해결한다.
ndarray1 = np.arange(start=1, stop=101)
print(ndarray1)

# # 각 요소 중 1의 자리수가 2인 수들만 추출하기
ndarray1_filtered = ndarray1[ndarray1 % 10 == 2]
print("결과:", ndarray1_filtered)
print("="*30)
#-----------------------------------------------------
# 1 ~ 49 요소 중 2와 5의 공배수 추출하기 

ndarray1 = np.arange(start=1, stop=50)
print("전체 배열:", ndarray1)

# 2와 5의 공배수를 추출
ndarray_filtered = ndarray1[(ndarray1 % 2 == 0) & (ndarray1 % 5 == 0)]

print("2와 5의 공배수:", ndarray_filtered)
print("="*30)
# 위에서 추출한 공배수 중 20 이상인 값만 추출하기 
ndarray_filtered1 = ndarray_filtered[ndarray_filtered >= 20]
print(ndarray_filtered1)
print("="*30)
#-----------------------------------------------------
# 데이터를 분석하여, 각 수치별 오름차순 및 내림차순 후 이름을 출력하세요.
# [[과일이름], [가격]], 가격별 내림차순
array1 = [['Mango', 'Apple', 'Pear', 'Pitch', 'Melon'], [1500, 1800, 2000, 2500, 8500]]

ndarray2 = np.array(array1)
fruit_name = ndarray2[:1].flatten()
fruit_price = ndarray2[-1]

print(fruit_name, fruit_price, sep="\n")

fruit_name_asc = fruit_name[fruit_price.argsort()[::-1]]
print(fruit_name_asc)
print("="*30)
#-----------------------------------------------------
# [[이름], [나이], [학점]], 나이별 내림차순, 학점별 오름차순
array2 = [['둘리', '또치', '길동', '도너', '마이콜'], [7, 9, 40, 5, 20], ['C', 'A', 'B', 'F', 'D']]

ndarray2 = np.array(array2)
names = ndarray2[:1].flatten()
ages = ndarray2[1].astype(np.int16)
grades = ndarray2[-1]

print(names, ages, grades)

print(names[ages.argsort()[::-1]])
print(names[grades.argsort()])
print("="*30)
#-----------------------------------------------------
# [[국가이름], [GDP 성장률]], GDP 성장률별 오름차순
array3 = [['한국', '영국', '러시아', '남아프리카 공화국', '미국'], ['1.7%', '-0.6%', '0.3%', '1.2%', '1.4%']]
gdp_list = []
ndarray2 = np.array(array3)

nations = ndarray2[0]
gdps = ndarray2[1]


for gdp in gdps:
    gdp_list.append(gdp.replace('%', ''))

gdps = np.array(gdp_list, dtype=np.float16)

print(nations[gdps.argsort()])
```
