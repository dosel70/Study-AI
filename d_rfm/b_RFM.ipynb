```python
# 무엇을 기준점으로 삼을 것인가 -> 결제 금액 
import numpy as np

Q1 = cp_df.describe().T.loc['TransactionAmount', '25%']
Q3 = cp_df.describe().T.loc['TransactionAmount', '75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
upper_bound = Q3 + iqr * 1.5

if lower_bound < 0:
    lower_bound = 0

print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
# 이상치를 제거 할 것인가 ? -> 일단 해보자 --> 이상치에 해당하는 값이 없다
delete_upper_df = cp_df[cp_df['TransactionAmount'] > upper_bound]
delete_upper_df

delete_lower_df = cp_df[cp_df['TransactionAmount'] < lower_bound]
delete_lower_df
```

```python
# feature명 변경 
cp_df = cp_df.rename(columns = {'ProductInformation':'Product'})
cp_df = cp_df.rename(columns = {'TransactionAmount':'Transaction'})
cp_df = cp_df.rename(columns = {'PurchaseDate' : 'date'})
cp_df = cp_df.rename(columns = {'OrderID' : 'Order'})
cp_df = cp_df.rename(columns = {'CustomerID' : 'Customer'})
cp_df
```

```python
display(cp_df)
```

```python
cp_df['Product'].value_counts()
```

```python
# 상품정보 value명을 변경 A,B,C,D
cp_df['Product'] = cp_df['Product'].replace(
    ['Product A', 'Product B', 'Product C', 'Product D'], ['A', 'B', 'C', 'D']
)

cp_df['Product'].value_counts()
```

```python
display(cp_df)
```

```python
# 상품정보 별로 금액을 묶어보자
# 상품 정보 별 평균 금액 
cp_df_a = cp_df.groupby('Product')['Transaction'].mean().reset_index()
display(cp_df_a)

# 상품 정보 별 최대 금액 
cp_df_b = cp_df.groupby('Product')['Transaction'].max().reset_index()
display(cp_df_b)

# 상품 정보 별 최소 금액 
cp_df_c = cp_df.groupby('Product')['Transaction'].min().reset_index()
display(cp_df_c)

# 상품 정보 별 금액 표준편차
cp_df_d = cp_df.groupby('Product')['Transaction'].std().reset_index()
display(cp_df_d)
```

#### 1차 분석 (상품정보 & 상품금액)
- 전반적으로 분석을 해보았을때, 평균적으로 가장 금액이 높은 상품은 A그룹이며, 그 다음으로 C, D, B 그룹순이다.
- 상품별 최대 금액은 C 그룹에서 999.44로 가장 높게 나왔으며 최소 금액은 D 그룹에서 12.13으로 가장 적은 최소값을 기록하였다.
- 그러나 한가지 의문점은 평균적으로 가장 가격이 높게 책정이 되었던 A 그룹에서 최소값 측정 결과 12.71로 하위 2번째로 최소값이
  낮았으며, 이를 미루어보았을 때, A그룹의 표준편차가 가장 클 것으로 예상이 되었다.
- 하지만 평균가격이 2순위였으며, 최대금액이 1순위 였던 C그룹의 표준편차가 가장 큰 것을 알 수 있었고, 아마 상품정보 말고도 가격과 연관된 정보가 있을 것으로 판단하여, 고객번호도 같이 연관을 지어서 분석을 하기로 하였다.


```python
# 고객번호의 빈도수를 Frequency로 엮어보자!
# 몇가지 문제점 -> 고객은 굉장히 데이터가 많을 텐데 이 고객정보를 어떻게 그룹으로 묶어서 Montery로 
# 만들 수 있을까 라는 의문점이 생겼다. -> 고객정보와 상품정보를 같이 frequency_array에 담아서 집계함수로 총합을 구해보자.
# 그 전에 우선 고객정보의 수는 매우 많고, 중복된 부분도 많기 때문에 우선 중복된 고객정보를 지우고 다 하나로 초기화하자
cp_df['Order'].value_counts()
```

```python
# 고객 정보는 중복된 행이 있기때문에 이를 제거한다.
cp_df = cp_df.drop_duplicates(subset=['Customer'])
cp_df['Customer'].value_counts()
```

```python
# grouped -> 지역별로 멤버를 담아놓은 변량 
grouped = cp_df.groupby('Location')['Customer'].apply(list).reset_index()
print(grouped)

# producted -> 상품별로 멤버를 담아놓은 변량 
producted = cp_df.groupby('Product')['Customer'].apply(list).reset_index()
print(producted)

# trancationed -> 고객별로 구매한 결제 금액 
trancationed = cp_df.groupby('Customer')['Transaction'].apply(list).reset_index()
print(trancationed)

# 상품 결제 최고 금액 (회원 아이디:9991, 결제 금액 : 999.44)
max_price = trancationed.sum()
print(max_price)

display(cp_df)
```

```python

# 각 고객의 총 구매 횟수 계산
cp_df['Monetary'] = cp_df.groupby('Customer')['Transaction'].sum().reset_index()
cp_df
## Montery 추출 완성 
```

```python
## Montery 이상치 조사 
import numpy as np

Q1 = cp_df.describe().T.loc['Monetary', '25%']
Q3 = cp_df.describe().T.loc['Monetary', '75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
uppper_bound = Q3 + iqr * 1.5

if lower_bound < 0:
    lower_bound = 0

print(f'정상치 범위: {lower_bound} ~ {uppper_bound}')

```

```python
mean_df_location = cp_df.groupby('Location')['Monetary'].mean().reset_index()
max_df_location = cp_df.groupby('Location')['Monetary'].max().reset_index()
min_df_location = cp_df.groupby('Location')['Monetary'].min().reset_index()
std_df_location = cp_df.groupby('Location')['Monetary'].std().reset_index()
display(mean_df_location)
display(max_df_location)
display(min_df_location)
display(std_df_location)
```
