#### DataFrame
- dict를 DataFrame으로 변환하고자 할 때 DataFrame 생성자에 전달한다.
- 컬럼명을 추가하거나 인덱스 명을 변경하는 등 다양하게 설정 할 수 있다.

```python
import pandas as pd
film = {
    'title': ['명량', '극한 직업', '범죄 도시3', '국제 시장'],
    'audience': [17_615_919, 16_266_480, 10_682_674, 14_265_222],
    'country': ['한국', '한국', '한국', '한국']
}
film_df = pd.DataFrame(film)
display(film_df)

# 새로운 컬럼명 
film_df.iloc[0]
film_df['income'] = [135_758_658_810, 139_657_105_516, 104_686_489_632, 110_951_970_230]
display(film_df)

# 인덱스명 변경
film_df.index = ['one','two','three','four']
display(film_df)

# 인덱스 초기화
film_df = film_df.reset_index()
display(film_df)

# feature 삭제 (열을 삭제해야함)
film_df = film_df.drop(labels=['index'], axis=1)
display(film_df)

# feature 이름 변경
film_df = film_df.rename(columns={"title":"name"})
display(film_df)

# 행 삭제 
film_df = film_df.drop(index=[2], axis=0)
display(film_df)

film_df = film_df.reset_index(drop=True) 
# 원본을 원래대로 바꾸려면 기존 index 값도 지워야하기 때문에 drop=True를 써줘야함!!
display(film_df)
```

#### describe()
- 숫자형 데이터의 개수, 평균, 표준편차, 최소값, 사분위 분포도(중앙값 : 50%), 최대값을 제공한다.
- 25 번째 백분위수와 75번째 백분위수를 기준으로 정상치의 범위를 설정할 수 있다.

```python
import pandas as pd
happiness_df = pd.read_csv('./datasets/happiness_report_2022.csv')
```

```python
display(happiness_df.describe().T)
```

```python
happiness_df.describe().T.loc['score','50%']
```

```python
import numpy as np
from scipy.stats import iqr

happiness_Q1 = np.percentile(happiness_df.score, 25) # 하위 25프로
happiness_Q3 = np.percentile(happiness_df.score, 75) # 상위 25프로
print(happiness_Q1, happiness_Q3)

happiness_Q1 = happiness_df.describe().T.loc['score','25%']
happiness_Q3 = happiness_df.describe().T.loc['score','75%']
print(happiness_Q1, happiness_Q3)

iqr_value = happiness_Q3 - happiness_Q1
print(iqr_value)

iqr_value = iqr(happiness_df.score)
print(iqr_value)

lower_bound = happiness_Q1 - 1.5 * iqr_value # 최소
upper_bound = happiness_Q3 + 1.5 * iqr_value # 최대 
print(f'정상치의 범위 : {lower_bound} ~ {upper_bound}')
```

```python

happiness_mean_df = happiness_df.groupby('income').mean('score').reset_index()
display(happiness_mean_df)

happiness_min_df = happiness_df.groupby('income').min('score').reset_index()
display(happiness_min_df)

happiness_max_df = happiness_df.groupby('income').max('score').reset_index()
display(happiness_max_df)

happiness_std_df = happiness_df.groupby('income')['score'].std().reset_index()
display(happiness_std_df)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 4, figsize=(15, 5))

happiness_mean_df.plot(title='mean', x='income', y='score', fontsize='4', kind='line', grid=True, ax=axes[0], ylabel="score")
happiness_min_df.plot(title='min', x='income', y='score', fontsize='4', kind='line', grid=True, ax=axes[1], ylabel="score")
happiness_max_df.plot(title='max', x='income', y='score', fontsize='4', kind='line', grid=True, ax=axes[2], ylabel="score")
happiness_std_df.plot(title='std', x='income', y='score', fontsize='4', kind='line', grid=True, ax=axes[3], ylabel="score")
```
